```python
# wep_scraping in WUZZUF
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import math
page = 0
job_title = []
company_name =[]
company = []
location = []
job_kind = []
skills = []

while True:
    response = requests.get(f"https://wuzzuf.net/search/jobs/?a=hpb&q=python&start={page}")
    #f"https://wuzzuf.net/search/jobs/?a=hpb&q=python&start={0}" 
    #python&start=0,1,2,3,4,5,6 (to last number 146 Jobs found / number of jobes in each page=15 )
    #("https://wuzzuf.net/search/jobs/?q=python&a=hpb")
    
    soup = BeautifulSoup(response.content, 'lxml')
    
    job_titles= soup.find_all('h2', {"class":"css-m604qf"})
    company_names =soup.find_all('a', {"class":"css-17s97q8"})
    locations = soup.find_all('span',{ "class":"css-5wys0k"})
    job_kinds = soup.find_all('span', {"class":"css-1ve4b75 eoyjyou0"})
    job_skills=soup.find_all('div', {"class":"css-y4udm8"})
    number_of_page = int(soup.find('strong').text)
    
    for job in range(len(job_titles)):
        job_title.append(job_titles[job].text)
        company_name.append(company_names[job].text.replace(' -',''))
        location.append(locations[job].text)  
        job_kind.append(job_kinds[job].text)
        skills.append(job_skills[job].text)

    page +=1
    if page >= (math.ceil(number_of_page/15)):
        break
```

```python
#dictionaries to build line by line and later convert to a DataFrame
df_dict={'job_title': job_title,'company_name':company_name,'location': location,'job_kind':job_kind,'skills':skills}
df = pd.DataFrame(df_dict)
df
```
